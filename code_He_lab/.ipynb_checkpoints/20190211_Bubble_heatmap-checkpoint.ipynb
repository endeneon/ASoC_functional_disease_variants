# Bubble heatmap

```sos
import pandas as pd, numpy as np
import os, math
from scipy.stats import norm
cwd = os.path.expanduser("~/Documents/GWAS_ATAC")
t = norm.ppf(0.975)
deci = 6
```

```sos
from collections import OrderedDict
gwas_dict = OrderedDict([("pgc2", "SZ (2014)"), ("bip", "BP (2018)"), ("mdd", "MDD (2018)"), ("asd", "ASD (2017)"), ("adhd", "ADHD (2017)"), 
                         ("intelligence", "Intelligence (2018)"), ("edu", "Education Attainment (2016)"), ("neuro", "Neuroticism (2018)"), 
                         ("height", "Height (2014)"), ("bmi", "BMI (2018)"), ("ibd", "IBD (2015)"), ("ad", "Alzheimer (2018)"), ("parkinson", "Parkinson (2019)")])
atac_seq_dict = OrderedDict([("Glut_FDR_0.05", "iN_Glut_ASoC"), ("NPC_FDR_0.05", "NPC_ASoC"), ("GA_all_peaks.narrowPeak.cleaned.hg19.merged", "iN_GABA_OCR"), 
                             ("DN_all_peaks.narrowPeak.cleaned.hg19.merged", "iN_DA_OCR"), ("CN_all_peaks.narrowPeak.cleaned.hg19.merged", "iN_Glut_OCR"), 
                             ("NSC_all_peaks.narrowPeak.cleaned.hg19.merged", "NPC_OCR"), ("ips_all_peaks.narrowPeak.cleaned.hg19.merged", "iPSC_OCR")])
general_dict = OrderedDict([("Coding_UCSC", "Coding"), ("Intron_UCSC", "Intron"), ("Promoter_UCSC", "Promoter"), ("Conserved_LindbladToh", "Conserved"), 
                            ("UTR_3_UCSC", "UTR_3"), ("UTR_5_UCSC", "UTR_5"), ("DHS_peaks_Trynka", "DHS_peaks"), ("Repressed_Hoffman", "Repressed")])
```

```sos
def get_torus_res(anno, gwas):
    res = pd.read_table(f"{cwd}/{gwas}_zscore/{anno}.{gwas}.zscore.torus", sep = "\s+", header = None, names = ["anno", "beta", "CI_low", "CI_high"])
    if res.shape[0] == 2:
        res = res.tail(1)
        res["beta"] = [math.log2(math.exp(x)) for x in res["beta"]]
        res["CI_low"] = [math.log2(math.exp(x)) for x in res["CI_low"]]
        res["CI_high"] = [math.log2(math.exp(x)) for x in res["CI_high"]]
        return res
    else:
        print (f"{anno} torus result for {gwas} GWAS dataset is not available.")
```

```sos
def get_anno_table(d):
    res = pd.DataFrame(columns = gwas_dict.keys())
    for anno in d.keys():
        index = d[anno]
        for gwas in gwas_dict.keys():
            beta = get_torus_res(anno, gwas).iloc[0, 1]
            low = get_torus_res(anno, gwas).iloc[0, 2]
            # check if zscore <= 0.5
            if norm.cdf(beta / ((beta - low) / t)) <= 0.5:
                res.loc[index, gwas] = np.around(beta, deci), np.around(-np.log10(norm.cdf(beta / ((beta - low) / t)) * 2), deci)
            else:
                res.loc[index, gwas] = np.around(beta, deci), np.around(-np.log10(norm.cdf(-beta / ((beta - low) / t)) * 2), deci)
    return res
```

```sos
## results are different, but should be identical. Do not use "1-" for large z score.
norm.cdf(-8.95567701846767), 1-norm.cdf(8.95567701846767)
```

```sos
get_anno_table(atac_seq_dict)
```

```sos
def get_anno_table_2(d):
    res = pd.DataFrame(columns = ["anno", "gwas", "logOR", "Pvalue", "sig"])
    for anno in d.keys():
        for gwas in gwas_dict.keys():
            beta = get_torus_res(anno, gwas).iloc[0, 1]
            low = get_torus_res(anno, gwas).iloc[0, 2]
            high = get_torus_res(anno, gwas).iloc[0, 3]
            sig = 1 if np.sign(low) == np.sign(high) else 0
            if norm.cdf(beta / ((beta - low) / t)) <= 0.5:
                res.loc[len(res)] = [d[anno], gwas_dict[gwas], np.around(beta, deci), norm.cdf(beta / ((beta - low) / t)) * 2, sig]
            else:
                res.loc[len(res)] = [d[anno], gwas_dict[gwas], np.around(beta, deci), norm.cdf(-beta / ((beta - low) / t)) * 2, sig]
    return res
```

```sos
atac_seq_table = get_anno_table_2(atac_seq_dict)
# general_table = get_anno_table_2(general_dict)
```
